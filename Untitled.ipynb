```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Dashboard de détection d’anomalies industrielles/src/merged_hourly_regional.csv')  # ton fichier
```

```python
print(df.head())
print(df.shape)
print(df.columns.tolist())
print(df.dtypes)
print(df.isnull().sum())
```

```python
df['date'] = pd.to_datetime(df['date'])  # adapte au nom de ta colonne
df = df.sort_values('date')
plt.figure(figsize=(15,4))
plt.plot(df['date'], df['conso_elec_mw'])  # adapte le nom
plt.title('Conso électrique France')
plt.xticks(rotation=45)
plt.show()
```
